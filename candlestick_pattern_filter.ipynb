```python
import pandas as pd 
import numpy as np 
import yfinance as yf
import datetime, time 
```

```python
# 蒐集每檔股票過去14個日曆天(非交易日)的「開、高、低、收、量」
def kline(ticker):
    start = datetime.date.today() - datetime.timedelta(days = 14)
    end = datetime.date.today()
    return pd.DataFrame(yf.download(ticker,start,end))

# 計時起始點
start = time.clock()

# 建立 2個空的 list，之後迴圈運算篩出強勢 & 弱勢股票可個自存放
long, short= [],[]

# 預製一個簡單的 csv檔，存放 S&P500的成分股，1欄是股票代號(必需)、1欄是公司名稱(非必需)
data = pd.read_csv('S&P500.csv')

tickers = { i: j for i,j in zip(list(data['Symbol']), list(data['Name']))}
for i in tickers.keys():
    # 用開頭的 kline函式，迴圈爬取csv檔裡每間公司近10筆資料(10個交易日)的開高低收量
    res = kline(str(i))[-10:]
    try:
        # 若符自己定義的股價強勢特徵，就存在「long」這個空 list裡
        if (len([res['Open'] > res['Adj Close'].shift()*1.02]) >= 1)&(len(res[res['Adj Close'] > res['Open']*1.03]) >= 1 )&(len( res[(res['High'] > res['High'].shift()) & ( res['Low'] > res['Low'].shift())] ) >= 5)&(res['Adj Close'][-1]*1.03 >= res['High'][-10:].max())&(res['Volume'][-3:].mean() > res['Volume'][-10:].mean()*1.1):
            long.append(i) 
        # 若符自己定義的股價弱勢特徵，就存在「short」這個空 list裡
        elif (len(res[res['Open'] < res['Adj Close'].shift()*0.98]) >= 1 )&(len(res[res['Adj Close']*1.03 < res['Open']]) >= 1 )&(len(res[ (res['High'] < res['High'].shift()) & (res['Low'] < res['Low'].shift()) ]) >= 4)&(res['Adj Close'][-1]*0.97<=res['Low'][-10:].min())&(len(res[res['Adj Close'] < res['Open']*0.98]) >= 4 )&(len( res [res['Adj Close'] < res['Open']] [res['High'] > res['Adj Close']*1.01] ) >= 1):
            short.append(i) 
        else:
            pass
    except Exception as e:
        # 若 csv檔裡某公司在模組 yfinance查不到時，就跳出提醒字串
        print('公司',tickers[i],'不存在' )
    continue

# 計時終止點，記錄全程運算耗時多久
end = time.clock() 
f'跑了{time.gmtime(end-start)[3]}時{time.gmtime(end-start)[4]}分{time.gmtime(end-start)[5]}秒'
```
